```python
import random
import string

TARGET = "Hello, World!"

POPULATION = 20
PROBABILITY_MUTATION = 0.5
```

```python
def generate_random_solution(length=13):
    return ''.join([random.choice(string.printable) for _ in range(length)])
```

```python
def evaluate(solution):
    def compute_error(s, t):
        return abs(ord(s) - ord(t))
    return sum([compute_error(s, t) for s, t in zip(solution, TARGET)])
```

```python
def mutate_solution(solution):
    index = random.randint(0, len(solution)-1)
    new_char = random.choice(string.printable)
    new_solution = solution[:index] + new_char +  solution[index+1:]
    return new_solution
```

```python
def cross_function(parent1, parent2):
    return parent1[:5] + parent2[5:]
```

```python
def sort_population(population):
    return sorted(population, key=lambda x: x[1])

population = []
for _ in range(POPULATION):
    best = generate_random_solution()
    best_score= evaluate(best)
    population.append((best, best_score))

population = sort_population(population)
```

```python
for _ in range(5000):
    
    best_element = population[0]
    best_string, best_score = best_element
    print(f'Best score so far {best_score}, Solution {best_string}')
    if best_score == 0:
        break
    
    new_population = population.copy()
    for _ in range(5):
        parent1, parent2 = random.choice(population)[0], random.choice(population)[0]
        child = cross_function(parent1, parent2)
        
        random_number = random.random()
        if random_number < PROBABILITY_MUTATION:
            child = mutate_solution(child)
        
        score_child = evaluate(child)
        new_population.append((child, score_child))
    new_population = sort_population(new_population)
    population = new_population[:POPULATION]
```
