```python
import random
import string

TARGET = "Hello, World!"
```

```python
def generate_random_solution(length=13):
    return ''.join([random.choice(string.printable) for _ in range(length)])
```

```python
def evaluate(solution):
    def compute_error(s, t):
        return abs(ord(s) - ord(t))
    return sum([compute_error(s, t) for s, t in zip(solution, TARGET)])
```

```python
def mutate_solution(solution):
    index = random.randint(0, len(solution)-1)
    new_char = random.choice(string.printable)
    new_solution = solution[:index] + new_char +  solution[index+1:]
    return new_solution
```

```python
best = generate_random_solution()
best_score= evaluate(best)
```

```python
for _ in range(5000):
    print('Best score so far', best_score, 'Solution', "".join(best))
    if best_score == 0:
        break

    new_solution = mutate_solution(best)

    new_score = evaluate(new_solution)
    if new_score < best_score:
        best, best_score = new_solution, new_score
```
